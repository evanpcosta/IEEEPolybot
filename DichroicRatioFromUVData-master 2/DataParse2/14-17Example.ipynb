```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#make sure all the data used to calculate the ratios are labeled with -90 or -0

data = pd.read_csv("/Users/evancosta/Desktop/DataParse2/14-17.csv") 
```

```python
data.shape
```

```python
data.columns[1]
```

```python

data = data.dropna(axis=1, how='all')
data = data.dropna()
data
```

```python
index = []
for x in data.columns:
    if 'Unnamed' in x:
       continue 
    else:
        index.append('alpha')
        index.append(x)
data.columns = index
data
```

```python
data = data.loc[:,~data.columns.duplicated()]
data
```

```python
data = data.drop(data.index[0])
data
```

```python
datarr = data.to_numpy(dtype=float)
datarr.shape
```

```python
for i in range(1, len(datarr[0, :])):
    plt.plot(datarr[:,i])
```

```python
change = 0
for i in range(1, len(datarr[0, :])):
    doublecheck = 0
    while doublecheck < 2:
        for x in range(len(datarr[:, i])-1):
            if change != 0:
                datarr[x][i] += change 
            else: 
                difference = datarr[x][i] - datarr[x+1][i]
                if abs(difference) > 0.01:
                    change = difference
        change = 0
        doublecheck += 1
for i in range(1, len(datarr[0, :])):
    plt.plot(datarr[:,i])
data.columns
```

```python

dellist = []
for i in range(4, len(datarr[0, :])):
    if '-0' in data.columns[i]:
        datarr[:,i] -= datarr[:, 3] 
    elif '-90' in data.columns[i]:
        datarr[:,i] -= datarr[:, 2]
    else:
        dellist.append(data.columns[i])
    

for x in dellist:
    count = 0
    for i in data.columns:
        if i == x:
            break
        count += 1
    datarr = np.delete(datarr, count, 1)
    data = data.drop(x, axis=1)


for i in range(4, len(datarr[0, :])):
    plt.plot(datarr[:,i])
print(dellist)
data.columns
```

```python
#make the minimum zero 
for i in range(1, len(datarr[0, :])):
    change = -min(datarr[:, i])
    for x in range(len(datarr[:, i])-1):
        datarr[x][i] += change
for i in range(4, len(datarr[0, :])):
    plt.plot(datarr[:,i])
```

```python
#remove the last row
datarr = datarr[:-1,:]
for i in range(4, len(datarr[0, :])):
    plt.plot(datarr[:,i])
data
```

```python
#compute ratio 
print(data.shape)
print(datarr.shape)
print(data.columns)

ratK1S = max(datarr[:, data.columns.get_loc('K1S-spin-90')])/max(datarr[:, data.columns.get_loc('K1S-spin-0')])
print(ratK1S)

ratff = max(datarr[:, data.columns.get_loc('K1S-ff-3mms-90')])/max(datarr[:, data.columns.get_loc('K1S-ff-3mms-0')])
print(ratff)

ratIf = max(datarr[:, data.columns.get_loc('K1S-lf-3mms-90')])/max(datarr[:, data.columns.get_loc('K1S-ff-3mms-0')])
print(ratIf)
```
